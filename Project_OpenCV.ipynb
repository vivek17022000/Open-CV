```python
import cv2
from google.colab.patches import cv2_imshow

def face_counter(img_path):     # Defining a function having a parameter for our image path/location.

  
  face_cascade = cv2.CascadeClassifier('/content/haarcascade_frontalface_default.xml')  # Loading Classifier from the given file.
  eye_cascade = cv2.CascadeClassifier('/content/haarcascade_eye.xml')

  img = cv2.imread(img_path)      # Loading the image from the given path.

  faces = face_cascade.detectMultiScale(img)    # Detecting faces from the given image.
  eyes = eye_cascade.detectMultiScale(img)      # Detecting eyes from the given image.

  count = 0  
  count1 = 0

  for (x , y , w, h) in faces:    # Getting the values(x,y,w,h) for each detected face.

    face_detected_img = cv2.rectangle(img, (x,y), (x+w, y+h), (240,0,19), 2)   # Drawing a rectangle on each detected face.
    count += 1     # Incrementing the value of count by 1 for each face detected.

  for (x, y, w, h) in eyes:       # Getting the values(x,y,w,h) for each detected eye.

    cv2.rectangle(img, (x,y), (x+w, y+h), (2,2,220), 2)   # Drawing a rectangle on each detected eye.
    count1 += 0.5   # Incrementing the value of count1 by 0.5 for each detected eye[will give 1 for each pair of eye]


  print('Number of faces detected : ', count)   # Printing the number of faces detected.

  print('Pair Of eyes Detected : ', round(count1))   # Printing the number of pairs of eyes detected.

  cv2_imshow(face_detected_img)   # Displaying the image with face and eyes detected.
```

```python
face_counter ('/content/images (1).jpg')
```
